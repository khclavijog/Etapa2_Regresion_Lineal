```python
# Se carga al entorno de trabajo las librerias necesarias

import matplotlib.pyplot as plt
from sklearn import linear_model
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
# Se carga la informacion del CSV a la variable data

data = pd.read_csv("C:/Users/haidy/Downloads/data.csv", sep=",")
```

```python
# Se muestran los datos anexados

data
```

```python
# Se realiza la grafica de dispersion

data.plot.scatter(x="metro", y="precio")
plt.show()
```

```python
# Se implementa la variable de regresion

regresion = linear_model.LinearRegression()
```

```python
# Se agregan los datos al vector

metro = data["metro"].values.reshape((-1,1))
```

```python
# Se crea el modelo

modelo = regresion.fit(metro, data["precio"])

#Interseccion

print("Interseccion (b)", modelo.intercept_)

#Pendiente

print("Pendiente (m)", modelo.coef_)
```

```python
# Definicion de las entradas a partir del modelo de prediccion

entrada= [[5],[15],[20],[25]]
predicciones = modelo.predict(entrada)
print(predicciones)
```

```python
# Grafica de Regresion

data.plot.scatter(x="metro", y="precio", label="Datos")
plt.scatter(entrada, predicciones, color="orange")
plt.plot(entrada, predicciones, color="black", label="Linea de regresion")
plt.xlabel('metro')
plt.ylabel('precio')
plt.legend()
plt.show()
```
